## Estudantes de Matemática


```python
# Verificar a distribuição de homens e mulheres
masc = len(math_students[math_students['sex'] == 'M'])
fem = len(math_students[math_students['sex'] == 'F'])
labels = ['Masculino','Feminino']
values = [masc,fem]
colors = ['#96D38C','#E1396C']
trace = go.Pie(labels=labels,values=values,
              hoverinfo='label+percent',textinfo='value',
              textfont=dict(size=20),
              marker=dict(colors=colors,
                         line=dict(color='#000000',width=2)))
ply.iplot([trace],filename='Estudantes de Matématica')
```

```python
# Distribuição da idade da turma de Matemática

masc_students = math_students[math_students['sex'] == 'M']
fem_students = math_students[math_students['sex'] == 'F']


masc_age = go.Bar(
    x=list(range(masc_students['age'].min(),masc_students['age'].max()+1)),
    y=list(masc_students['age'].value_counts().sort_index()),
    name='Masculino',
    marker=dict(
        color='rgb(49,130,189)'
    )   
)

fem_age = go.Bar(
    x=list(range(fem_students['age'].min(),fem_students['age'].max()+1)),
    y=list(fem_students['age'].value_counts().sort_index()),
    name='Feminino',
    marker=dict(
        color='rgb(204,204,204)',
    )   
)

data = [masc_age, fem_age]
layout = go.Layout(
    xaxis=dict(tickangle=-45),
    barmode='group',
)

fig = go.Figure(data=data, layout=layout)
ply.iplot(fig, filename='angled-text-bar')
```

# Estudantes de Matemática e Português


```python
# Análisar o consumo de bebida álcoolica por idade dos meninos e das meninas nos dias de trabalho 
stud_math_port = pd.merge(port_students,math_students,how='outer')
df_age_teen = stud_math_port[(stud_math_port['age'] == 15) | (stud_math_port['age'] == 16) | (stud_math_port['age'] == 17)]
df_age_teen['Dalc'].value_counts().sort_index()

df_age_young = stud_math_port[(stud_math_port['age'] == 18) | (stud_math_port['age'] == 19) | (stud_math_port['age'] == 20) | (stud_math_port['age'] == 21) | (stud_math_port['age'] == 22)]
df_age_young['Dalc'].value_counts().sort_index()

trace1 = {
  "x": list(df_age_teen['Dalc'].value_counts().sort_index()), 
  "y": ["Very Low", "Low", "Moderate", "High", "Very High"], 
  "name": "Age: 15-17", 
  "orientation": "h", 
  "type": "bar"
}
trace2 = {
  "x": list(df_age_young['Dalc'].value_counts().sort_index()), 
  "y": ["Very Low", "Low", "Moderate", "High", "Very High"], 
  "name": "Age: 18-22", 
  "orientation": "h", 
  "type": "bar"
}

data = go.Data([trace1, trace2])
layout = {
  "barmode": "group", 
  "title": "Student Consumption Workday alcohol "
}
fig = go.Figure(data=data, layout=layout)
ply.iplot(fig, filename='angled-text-bar')

#df[(df['C']==1) | (df['B']==True)]
#stud_math_port.groupby('Dalc').count()
```
